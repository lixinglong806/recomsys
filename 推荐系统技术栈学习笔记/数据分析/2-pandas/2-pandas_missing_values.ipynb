```python
# 数据清洗
import pandas as pd

from pandas import Series,DataFrame

import numpy as np
```

```python
df = DataFrame(np.random.randint(0,150,size=(150,3)),columns=['Python','Math','En'])
df
```

```python
for i in range(15):
    index = np.random.randint(0,150,size=1)[0]
    columns = np.random.randint(0,3,size=1)[0]
    df.iloc[index,columns] = None
```

```python
# 非空数据
cond = df.notnull().all(axis=1)
df[cond]
```

```python
# na-空数据，dropna删除空数据
df.dropna()
```

```python
# 删除列
df.drop(labels=['En'],axis=1)
```

```python
df.isnull().any(axis=1)
```

```python
# 有空值的行index
cond = df.isnull().any(axis=1)
index = df[cond].index
index
```

```python
# 删除有空值的行，得到无空值的行
df2 = df.drop(labels=index)
df2
```

```python
# 得分小于60的，删除
cond = (df2 < 60).any(axis = 1)
index = df2[cond].index
df2.drop(labels = index)
```

```python
# 注意以下两种index的结果不同
(df2 < 60).any(axis = 1).index
```

```python
cond = (df2 < 60).any(axis = 1)
df2[cond].index
```

```python
# 平均分小于90进行过滤
df2
```

```python
cond = df2.mean(axis = 1)<90
index = df2[cond].index
df2.drop(labels = index)
```

```python
# 小于60，大于100分
cond1 = df2.mean(axis = 1)<60
cond1
```

```python
cond2 = df2.mean(axis = 1)>100
cond2
```

```python
cond = cond1|cond2
df2[cond]
```

填充函数，填充的原则：使数据更合理

```python
df
```

```python
df.mean()
```

```python
df.fillna(60)
```

```python
# 每列的平均值 填充到每列的空值去
df.fillna(value = df.mean())
```

```python
df.median()
```

```python
#填充中位数，
df.fillna(value=df.median())
```

```python
#向前/后填充
df
```

```python
'''method : {'backfill', 'bfill', 'pad', 'ffill', None}, default None
    Method to use for filling holes in reindexed Series
    pad / ffill: propagate last valid observation forward to next valid
    backfill / bfill: use next valid observation to fill gap.'''
df.fillna(method='backfill')
```

```python
# NaN前无值则无法填充空值
df.fillna(method='pad')
```

```python
# axis
df.fillna(method='backfill',axis=1)
```
