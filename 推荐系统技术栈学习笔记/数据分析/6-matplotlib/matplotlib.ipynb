### 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x=np.linspace(-2,2,100)

y=np.linspace(-2,2,100)


z=(x**2 + y**2)**0.5

fig=plt.figure(figsize=(12,9))

axes3D=Axes3D(fig)

axes3D.plot(x,y,z)

X,Y=np.meshgrid(x,y)
Z=(X**2+Y**2)**0.5
axes3D.plot_surface(X,Y,Z)
```
