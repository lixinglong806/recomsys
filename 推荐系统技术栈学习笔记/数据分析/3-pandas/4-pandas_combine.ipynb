2.使用pd.merge()合并

```python
#一对一合并
df1=DataFrame({'id':[1,10,1024],'name':['po','alisa','Michael'],'sex':['male','female','male']})
df1
```

```python
df2=DataFrame({'id':[1,11,1025],'salary':[10000,20000,30000],'age':[20,30,40]})
df2
```

```python
pd.concat([df1,df2])
```

```python
# 融合
df1.merge(df2)
```

```python
# 多对一，df1的id与df2的id
df2 = DataFrame({'id':[1,1,1],'salary':[10000,20000,30000],'age':[20,30,40]})
df2
```

```python
df1.merge(df2)
```

```python
# 多对多
df3=DataFrame({'id':[1,1,1],'name':['po','alisa','Michael'],'sex':['male','female','male']})
df3
```

```python
# 3*3
pd.merge(df2,df3)
```

4)key的规范化

```python
df1=DataFrame({'id':[1,10,1024],'name':['po','alisa','Michael'],'sex':['male','female','male']})
df1
df2=DataFrame({'id':[1,11,1025],'name':['softpo','alisa','Michael'],'age':[20,30,40]})
df2
display(df1,df2)
```

```python
# name不一致
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='id')
```

```python
df1
```

```python
df2 = DataFrame({'Id':[1,11,1025],'salary':[10000,20000,30000],'age':[20,30,40]})
df2
```

```python
# 列索引没有一个是相同的
display(df1,df2)
```

```python
# 列索引没有一个是相同的
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,left_on='id',right_on='Id')
```

5)内合并与外合并

```python
'''how : {'left', 'right', 'outer', 'inner'}, default 'inner' '''
pd.merge(df1,df2,how='outer',left_on='id',right_on='Id')
```

```python
# 左合并，左边所有数据都要保留
pd.merge(df1,df2,how='left',left_on='id',right_on='Id')
```

6)列冲突的解决

```python
df1=DataFrame({'id':[1,10,1024],'name':['po','alisa','Michael'],'sex':['male','female','male']})
df1
df2=DataFrame({'id':[1,11,1025],'name':['softpo','alisa','Michael'],'age':[20,30,40]})
df2
display(df1,df2)
```

```python
# suffixes:后缀
pd.merge(df1,df2,on='id',suffixes=('_A','_B'))
```

```python
df=DataFrame(np.random.randint(0,150,size=(10,3)),columns=['Python','Math','En'],index=list('ABCDEFHIJK'))
df
```

```python
s=df.sum(axis=1)
s
```

```python
# s是Series，不能和DataFrame融合
pd.merge(df,s)
```

```python
#将s转换为DataFrame
df_sum=DataFrame(df.sum(axis=1),columns=['sum'])
df_sum
```

```python
pd.merge(df,df_sum,left_index=True,right_index=True)
```
