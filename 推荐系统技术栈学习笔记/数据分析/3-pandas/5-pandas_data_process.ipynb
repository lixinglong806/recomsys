5.数据聚合【重点】
<br>数据聚合是数据处理的最后一步，通常是要使每一个数组生成一个单一的数值。
<br>
<br>数据分类处理：
<br>    分组：先把数据分成几组
<br>    用函数处理：为不同组的数据应用不同的函数以转换数据
<br>    合并：把不同组得到的结果合并起来
<br>
<br>数据分类处理的核心： groupby()函数

```python
df = DataFrame({'item':np.random.randint(0,10,size=100),
          'sailer':np.random.randint(0,10,size=100),
          'weight':np.random.randint(30,300,size=100),
          'price':np.random.randint(1,20,size=100)})
df
```

```python
def convert_item(x):
    if x<2:
        return '萝卜'
    elif x<7:
        return '白菜'
    else:
        return '青椒'
df['item'] = df['item'].map(convert_item)
```

```python
df
```

```python
def convert_sailer(x):
    if x<5:
        return '张大妈'
    elif x<7:
        return '李大妈'
    else:
        return '赵大爷'

df['sailer']=df['sailer'].map(convert_sailer)
```

```python
df
```

```python
def convert_weight(x):
    if x<80:
        return 50
    elif x<150:
        return 100
    else:
        return 150
    
df['weight']=df['weight'].transform(convert_weight)
```

```python
df
```

```python
# 统计这一个月，张大妈，李大妈，赵大爷，销售了多少斤蔬菜
ret = df.groupby(['sailer'])['weight']
ret
```

```python
# 分好组了，没有进行计算，销售总额，就是求和
ret.sum()
```

```python
ret.min()
```

```python
ret.apply(np.sum)
```

```python
# 算次数
ret.count()
```

```python
def count(x):
    return x.size
```

```python
ret.apply(count)
```

```python
def analysis(x):
    return (x.size,x.sum(),np.round(x.mean(),decimals=2),x.min())

ret.apply(analysis)
```
