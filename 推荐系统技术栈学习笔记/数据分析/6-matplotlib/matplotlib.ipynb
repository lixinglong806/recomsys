### 二、设置plot的风格和样式

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

#### 颜色

```python
x = np.linspace(0,2*np.pi,50)
```

```python
plt.plot(x,np.sin(x),color='red')
```

```python
# 16进制数表示颜色，'#ffee00'表示0xff ee 00
#三段分别表示红绿蓝
#jpg 0-255;   png  0-1
```

```python
plt.plot(x,np.sin(x),c='#ffff00')
```

```python
# 颜色还可使用归一化到(0,1]的RGB元组
# color = (0.3,0.3,0.4)
plt.plot(x,np.sin(x),c=np.random.rand(3))
```

#### 透明度
alpha参数

```python
plt.plot(x,np.sin(x),alpha=0.2)
```

#### 背景色
设置背景色，通过plt.subplot()方法传入facecolor参数，来设置坐标轴的背景色

```python
ax = plt.subplot(1,1,1,facecolor='green')

ax.plot(x,np.sin(x),color='r')
```

#### 线性
'-'   '--'   '-.'   ':'    'steps'    'None'/','

```python
plt.plot(x,np.sin(x),linestyle='steps')
```

```python
plt.plot(x,np.sin(x),ls='steps')
```

#### 线宽
linewidth或lw参数

```python
plt.plot(x,np.sin(x),linestyle='steps',lw=5)
```

#### 不同宽度的破折线
dashes参数
<br>设置破折号序列各段的宽度

```python
# dashes里的数字：1-线宽；2-间距；5-线宽；3-间距；2-线宽；5-间距
plt.plot(x,np.sin(x),dashes=[1,2,5,3,2,5])
```

marker参数

```python
plt.plot(x,np.sin(x),marker = '1',markersize=20)
```

```python
plt.plot(x,np.sin(x),marker = 's')
```

```python
plt.figure(figsize=(12,9))
plt.plot(x,np.sin(x),marker = '*',markersize=20)
```

```python
plt.plot(x,np.sin(x),marker = '|',markersize=20)
```

#### 多参数使用

```python
plt.plot(x,np.sin(x),c='r',marker = '*',markersize=15,markerfacecolor='g',markeredgecolor='b')
```

#### 在一条语句中为多个曲线进行设置
##### 多个曲线统一设置
属性名声明
<br>plt.plot(x1,y1,x2,y2,fmt,...)

```python
plt.plot(x,np.sin(x),x,np.cos(x),c='r',alpha=0.5)
```

#### 多个曲线不同设置
多个都进行设置时，无需声明属性plt.plot(x1,y1,fmt1,x2,y2,fmt2,...)

```python
plt.plot(x,np.sin(x),'r--o',x,np.cos(x),'g:*')
```

#### 三种设置方式

#### 向方法传入关键字参数

```python
plt.plot(x,np.sin(x),c='r')
```

#### 对实例使用一系列的setter()方法

```python
line, = plt.plot(x,np.sin(x))

line.set_color('#ff00ff')

line.set_ls('--')
```

#### 使用setp()方法

```python
line1,line2 = plt.plot(x,np.sin(x),x,np.cos(x))

#plt 是模块方法
plt.setp(line1,color='r',ls='steps')
```

#### X、Y轴坐标刻度
xticks()和yticks()方法

```python
plt.plot(x,np.sin(x))
#  若要使图上面的那几行不出现，则使用_=plt.xticks(np.arange(7),list('abcdefg'))
plt.xticks(np.arange(7),list('abcdefg'))
#fontsize字体大小
plt.yticks([-1,0,1],['min',0,'max'],c='r',fontsize=20,rotation=60)
```

```python
# 希腊字母

plt.plot(x,np.sin(x))
#  若要使图上面的那几行不出现，则使用_=plt.xticks(np.arange(7),list('abcdefg'))
plt.xticks(np.arange(0,2*np.pi+0.01,np.pi/2),[0,'$\pi/2$','$\pi/2$','$\pi$','$3\pi/2$','$2\pi/2$'],fontsize=30)
#fontsize字体大小
plt.yticks([-1,0,1],['min',0,'max'],c='r',fontsize=20,rotation=60)
```

### 五、3D图形（自学）

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x=np.linspace(-2,2,100)

y=np.linspace(-2,2,100)


z=(x**2 + y**2)**0.5

fig=plt.figure(figsize=(12,9))

axes3D=Axes3D(fig)

axes3D.plot(x,y,z)

X,Y=np.meshgrid(x,y)
Z=(X**2+Y**2)**0.5
axes3D.plot_surface(X,Y,Z)
```
